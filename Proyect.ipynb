```python
import networkx as nx
import matplotlib.pyplot as plt

class PlayerGraph:
    def __init__(self, players):
        self.players = players
        self.graph = []

    def add_edge(self, player1, player2, distance):
        self.graph.append([player1, player2, distance])

    def find(self, parent, i):
        if parent[i] == i:
            return i
        return self.find(parent, parent[i])

    def apply_union(self, parent, rank, x, y):
        xroot = self.find(parent, x)
        yroot = self.find(parent, y)
        if rank[xroot] < rank[yroot]:
            parent[xroot] = yroot
        elif rank[xroot] > rank[yroot]:
            parent[yroot] = xroot
        else:
            parent[yroot] = xroot
            rank[xroot] += 1

    def kruskal_algo(self, player_to_compare):
        result = []
        i, e, total_distance = 0, 0, 0
        self.graph = sorted(self.graph, key=lambda item: item[2])
        parent = [i for i in range(len(self.players))]
        rank = [0] * len(self.players)

        while e < len(self.players) - 1:
            player1, player2, distance = self.graph[i]
            i += 1
            x = self.find(parent, player1)
            y = self.find(parent, player2)

            if x != y:
                e += 1
                result.append([player1, player2, distance])
                self.apply_union(parent, rank, x, y)

        for player1, player2, dist in result:
            total_distance += dist
            print(f"{self.players[player1]} - {self.players[player2]}: {dist}")

        print("Costo total del MST para el grafo: ", total_distance)

        return result

player_graph = PlayerGraph(player)

player_to_compare = input("Ingrese el nombre del jugador a comparar: ")

try:
    index_to_compare = player.index(player_to_compare)
except ValueError:
    print("Jugador no encontrado.")
    exit()

for i in range(len(player)):
    if i != index_to_compare:
        distance = abs(goal[index_to_compare] - goal[i]) + abs(assist[index_to_compare] - assist[i]) + abs(dribbling[index_to_compare] - dribbling[i]) + abs(tackle[index_to_compare] - tackle[i])
        player_graph.add_edge(index_to_compare, i, distance)

result_edges = player_graph.kruskal_algo(index_to_compare)

G = nx.DiGraph()

for i in range(len(player)):
    G.add_node(player[i])

for edge in result_edges:
    G.add_edge(player[edge[0]], player[edge[1]])

pos = nx.spring_layout(G)
labels = {player[i]: player[i] for i in range(len(player))}
nx.draw(G, pos, with_labels=True, labels=labels, font_size=8, font_color='black', node_size=700, font_weight='bold', node_color='lightblue', edge_color='gray')
plt.title(f'Árbol de Expansión Mínima para {player_to_compare}')
plt.show()






distances = [(i, abs(goal[index_to_compare] - goal[i]) + abs(assist[index_to_compare] - assist[i]) + abs(dribbling[index_to_compare] - dribbling[i]) + abs(tackle[index_to_compare] - tackle[i])) for i in range(len(player)) if i != index_to_compare]


distances.sort(key=lambda x: x[1])


top_5_players = distances[:5]


G_top_5_players = nx.DiGraph()


for i in range(len(top_5_players)):
    G_top_5_players.add_node(player[top_5_players[i][0]])


for i in range(len(top_5_players)):
    distance = top_5_players[i][1]
    G_top_5_players.add_edge(player[index_to_compare], player[top_5_players[i][0]], weight=distance)


pos_top_5_players = nx.spring_layout(G_top_5_players)


node_colors = ['lightblue' if node != player[index_to_compare] else 'red' for node in G_top_5_players.nodes]


edge_labels = {(player[index_to_compare], player[top_5_players[i][0]]): f"{distance:.2f}" for i, (index, distance) in enumerate(top_5_players)}

nx.draw(G_top_5_players, pos_top_5_players, with_labels=True, labels=labels_top_5_players, font_size=8, font_color='black', node_size=700, font_weight='bold', node_color=node_colors, edge_color='gray')
nx.draw_networkx_edge_labels(G_top_5_players, pos_top_5_players, edge_labels=edge_labels, font_color='black')
plt.title(f'Árbol de Expansión Mínima para {player_to_compare} - 5 Jugadores con Menor Distancia')
plt.show()

```
